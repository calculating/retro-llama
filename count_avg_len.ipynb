```python
import json

# load links from links.json
with open('links.json', 'r') as f:
    links = json.load(f)

# get number of characters, then plot a histogram
import matplotlib.pyplot as plt
import numpy as np

# get number of characters
num_chars = [len(link) for link in links]

# plot histogram
plt.hist(num_chars, bins=50)
plt.xlabel('Number of characters')
plt.ylabel('Number of links')
plt.title('Distribution of number of characters in links')
plt.show()
```
